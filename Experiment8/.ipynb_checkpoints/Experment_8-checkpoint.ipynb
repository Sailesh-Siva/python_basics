```python
class ITRaid(Exception):
    def __init__(self, errmsg=""):
        self.errmsg = errmsg

    def __str__(self):
        return f"ITRaid Exception: {self.errmsg}"


class Politician:
    def getLIMIT(self):
        return 5 * 10**7

    def doITRaid(self):
        if self.hasProperty():
            return self.evaluateProperty()
        return 0


class Professional:
    def getLIMIT(self):
        return 10 * self.monthly_income * 12

    def doITRaid(self):
        if self.hasDeposit():
            return self.getDeposit()
        return 0


class Person:
    def __init__(self, name, status):
        self.__name = name
        self.__status = status

    def getStatus(self):
        return self.__status

    def ITRaid(self):
        return self.doITRaid()


class Minister(Person, Politician):
    def __init__(self, name, status, party_name, asset_value):
        super().__init__(name, status)
        self.party_name = party_name
        self.asset_value = asset_value

    def hasProperty(self):
        return bool(self.asset_value)

    def evaluateProperty(self):
        return sum(self.asset_value.values())


class Officer(Person, Professional):
    def __init__(self, name, status, qualification, monthly_income, deposit_value):
        super().__init__(name, status)
        self.qualification = qualification
        self.monthly_income = monthly_income
        self.deposit_value = deposit_value

    def hasDeposit(self):
        return bool(self.deposit_value)

    def getDeposit(self):
        return sum(self.deposit_value.values())


def perform_raids(person_list):
    for person in person_list:
        raid_value = person.ITRaid()
        limit = person.getLIMIT()
        if raid_value > limit:
            if isinstance(person, Minister):
                raise ITRaid("You have Disproportionate Assets")
            elif isinstance(person, Officer):
                raise ITRaid("Your Deposits are Locked")
        else:
            print("Good")


person_list = [
    Minister("Minister A", "Minister", "PartyX", {"Property1": 4 * 10**7, "Property2": 2 * 10**7}),
    Officer("Officer B", "Officer", "MSc", 50000, {"Account1": 2 * 10**6, "Account2": 1 * 10**6}),
    Minister("Minister C", "Minister", "PartyY", {"Property1": 3 * 10**7}),
    Officer("Officer D", "Officer", "PhD", 70000, {"Account1": 3 * 10**6, "Account2": 5 * 10**6}),
]

try:
    perform_raids(person_list)
except ITRaid as e:
    print(e)
```
